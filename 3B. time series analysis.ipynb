# Convert the object to datetime format

The format depends on your data. Here are a few examples

    1/12/2001: %d/%m/%Y
    2001/12/1: %Y/%m/%d
    2001-12-01: %Y-%m-%d
    2001-12-01 09:00: %Y-%m-%d %H:%M

```python
data['Date'] = pd.to_datetime(data['Date'], format='%Y-%m-%d %H:%M:%S')
```

# Missing values

```python
data = pd.read_csv("Nifty_daily_close_2023.csv")
```

```python
data_missing = pd.read_csv("Nifty_daily_close_2023_missing.csv")
```

```python
data_missing.head(10)
```

```python
plt.figure(figsize=(12,6))

# plt.plot(data.index, data['Close'], label='NIFTY CLosing Prices')
plt.plot(data_missing.index, data_missing['Close'], label='NIFTY CLosing Prices')
plt.legend(loc='best')
plt.show()
```

```python
plt.figure(figsize=(12,6))

plt.plot(data.index, data['Close'], label='NIFTY CLosing Prices')
plt.plot(data_missing.index, data_missing['Close'].ffill(), label='NIFTY CLosing Prices_filled')
plt.legend(loc='best')
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(data.index, data['Close'], label='NIFTY CLosing Prices')
plt.plot(data_missing.index, data_missing['Close'].bfill(), label='NIFTY CLosing Prices_filled')
plt.legend(loc='best')
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(data.index, data['Close'], label='NIFTY CLosing Prices')
plt.plot(data_missing.index, data_missing['Close'].interpolate(), '-o', markersize=3, label='NIFTY CLosing Prices_filled')
plt.legend(loc='best')
plt.show()
```

```python
fig = px.line(data_frame=data, x=data.index, y=['Close'], height=500)
fig.add_scatter(x=data.index, y=data_missing['Close'].interpolate(method='linear'), name='filled')
fig.show()
```

```python
fig = px.line(data_frame=data, x=data.index, y=['Close'], height=500)
fig.add_scatter(x=data.index, y=data_missing['Close'].bfill(), mode='lines+markers', name='filled')
fig.show()
```
